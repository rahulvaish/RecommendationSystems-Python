We have seen **SUPPORT** , **CONFIDENCE** and **LIFT**. Let's go witha practical example:

```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
data= pd.read_csv('D://DataScienceCollection//DataSets//apriori.csv')
```

```python
data.head(5)
```

```python
data.shape
```

#### **NOTE:** Computing 7500 rows, is computationally very expensive. Therefore we choose only top 2000 rows

```python
someData = data.head(2000)
```

```python
someData
```

```python
records =[]
for i in range(0,1999):
    records.append([str(someData.values[i,j]) for j in range(0,20)])
```

```python
association_rules = apriori(records, min_support=0.0056, min_confidence=0.2, min_lift=3, min_length=3)
association_results = list(association_rules)
```
